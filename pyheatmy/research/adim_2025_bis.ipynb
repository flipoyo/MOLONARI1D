# Problème général
On cherche à solutionner le couple d'équations suivantes:

$$
\left\{\begin{array}{lll} 
S_s \partial_t h & = & \partial_z (K \partial_z H) + q_{l} \\
\rho_m c_m \partial_t T & = & \rho_w c_w q \partial_z T + \lambda_m \partial_z^2 T 
\end{array} \right.
\quad \text{avec} \quad  q = K \partial_z H \, \, ; \quad  K = \frac{k\rho_w g}{\mu_w} = \frac{k g}{\nu_w} \, \, ; \quad  q_{l} = \text{flux latéral } (s^{-1})
$$
L'adimensionnement de ces l'équation est utile pour simplifier la résolution si certains termes sont dominants. Ce n'est pas le cas ici : les termes peuvent tous être importants, cela dépend du régime.
En outre, en adimensionnant correctement l'équation, tout les paramètres adimensionnés sont de l'ordre de l'unité. Cela limite donc les erreurs de calcul. Cela augmente aussi l'efficacité de calcul.

On notera que l'axe z est dirigé vers le bas, d'où la définition de ${\Delta T}$ et ${\Delta H}$ par la suite. 
## Adimensionnement et premières déductions
On pose les changements de variables suivants :

$$
\tilde{z}= \frac{z}{h} \quad \text{;} \quad \tilde{t}= \frac{t}{P} \quad \text{;} \quad \tilde{H} = \frac{H - H_{riv}}{H_{nap}-H_{riv}} =  \frac{H - H_{riv}}{\Delta H} \quad \text{;} \quad \tilde{\theta} = \frac{\theta - T_{riv}}{T_{nap}-T_{riv}} =  \frac{\theta - T_{riv}}{\Delta T} 
$$

Ici $h$ est la prfondeur de la colonne de sol étudiée, $P$ est la période caractéristique de variations de la température.

On exprime les équations avec les nouvelles variables pour obtenir les équations adimensionnées. La perméabilité $K$, l'emmagasinement spécifiques $S_{s}$ et le flux latéral $q_{l}$ sont pris constants à l'intérieur d'une couche, domaine de validité de ces équation :


$$
\left\{\begin{array}{rll} 
\dfrac{S_s \Delta H }{P} \, \, \partial_{\tilde{t}} \tilde{H} & = & \dfrac{K \Delta H}{h^2}  \, \, \partial_{\tilde{z}}^2 \tilde{H} \, \, + q_{l} \\
\dfrac{\rho_m c_m \Delta T}{P}  \, \, \partial_{\tilde{t}} \tilde{\theta}  & = & \dfrac{\rho_w c_w K \Delta H \Delta T}{h^2}  \, \, \partial_{\tilde{z}} \tilde{H}  \, \, \partial_{\tilde{z}} \tilde{\theta} + \dfrac{\lambda_m \Delta T}{h^2}  \, \, \partial_{\tilde{z}}^2  \tilde{\theta} 
\end{array} \right.
\quad \text{avec} \quad  K = \frac{k\rho_w g}{\mu_w} = \frac{k g}{\nu_w}
$$


On regroupe ensuite les paramètres pour obtenir les composantes principales du système:


$$
\left\{\begin{array}{rll} 
\overbrace{\dfrac{S_s h^2}{P K }}^{\large{\gamma}}  \, \, \partial_{\tilde{t}} \tilde{H} & = & \partial_{\tilde{z}}^2 \tilde{H} \, \,
+ \overbrace{\dfrac{h^2 q_{l}}{K \Delta H}}^{\large{\zeta}} \\[15pt]
\underbrace{\dfrac{\rho_m c_m h^2}{P \lambda_m}}_{\large{\beta}}  \, \, \partial_{\tilde{t}} \tilde{\theta}  & = & \underbrace{\dfrac{\rho_w c_w K \Delta H}{\lambda_m}}_{\large{\kappa}}  \, \, \partial_{\tilde{z}} \tilde{H} \, \, \partial_{\tilde{z}} \tilde{\theta} + \partial_{\tilde{z}}^2  \tilde{\theta} 
\end{array} \right.
\quad \text{avec} \quad  K = \frac{k\rho_w g}{\mu_w} = \frac{k g}{\nu_w}
$$

On considère donc le système adimensionné suivant:


$$
\left\{\begin{array}{rll} 
\gamma \times \partial_{\tilde{t}} \tilde{H}      & = & \, \zeta & + & \partial_{\tilde{z}}^2 \tilde{H} \\
\beta  \times \, \, \partial_{\tilde{t}} \tilde{\theta} & = & \kappa \times \partial_{\tilde{z}} \tilde{H} \times \partial_{\tilde{z}} \tilde{\theta} & + & \partial_{\tilde{z}}^2 \tilde{\theta} 
\end{array} \right.
$$

avec les 4 paramètres (sans dimensions) calculés dans la couche :
$$
\kappa = \frac{\rho_w c_w k g \Delta H}{\mu_w \lambda_m} \quad ; \quad \beta = \frac{\rho_m c_m h^2}{\lambda_m P }  \quad ; \quad  \gamma = \frac{\mu_w h^2 S_s}{k \rho_w g P} \quad ; \quad \zeta = \dfrac{h^2 q_{l}}{K \Delta H}
$$

IMPORTANT: Cette analyse montre que les profils de température et de charge dépendant uniquement de ces quatre paramètres $\kappa$, $\beta$, $\gamma$ et $\zeta$. Il est donc possible (sauf cas particulier) de déterminer ces quatre paramètres si on nous donne un profil.

Le problème initial contient lui plus de paramètres: $n$, $k$, $S_s$, $\lambda_m$, $\rho$ et $c$. Il est impossible de remonter à tout ces paramètres. C'est facile à voir pour $\rho$ et $c$: en effet, dans les équations, seul le produit $\rho c$ appraît il est donc impossible de remonter indépendamment à $\rho$ et $c$.
Le fait que le profils ne dépende que du quadriplet ($\kappa$, $\beta$, $\gamma$, $\zeta$) nous montre qu'il est impossible de remonter au 6-uplet ($n$, $k$, $S_s$, $\lambda_m$, $\rho$, $c$). En effet plusieurs 6-uplets sont équivalent acr ils donnent le même quadriplet ($\kappa$, $\beta$, $\gamma$, $\zeta$). Il y a donc une infinité de solutions, exactes, au problème d'inversion.

## 1. Discrétisation de l'équation de la charge hydraulique ($\tilde{H}$)

Nous commençons par l'équation de la charge hydraulique, qui est une équation de diffusion-source.

### 1.1. Rappel de l'équation adimensionnelle

L'équation à résoudre est la suivante :
$$
\gamma \cdot \frac{\partial \tilde{H}}{\partial \tilde{t}} = \frac{\partial^2 \tilde{H}}{\partial \tilde{z}^2} + \zeta
$$

### 1.2. Discrétisation spatio-temporelle

Nous discrétisons le domaine continu en une grille de points :
- **Espace :** L'axe $\tilde{z} \in [0, 1]$ est divisé en $N$ cellules de taille $\Delta\tilde{z} = 1/N$. L'indice spatial est noté $k \in \{0, ..., N-1\}$.
- **Temps :** Le temps progresse par pas de $\Delta\tilde{t}$. L'indice temporel est noté $j$.

La variable discrétisée est $\tilde{H}_k^j$, représentant la charge au nœud $k$ et au temps $j$.

### 1.3. Approximation par différences finies

Les dérivées sont approximées par des différences finies :
- **Dérivée temporelle (ordre 1) :**
  $$
  \frac{\partial \tilde{H}}{\partial \tilde{t}} \approx \frac{\tilde{H}_k^{j+1} - \tilde{H}_k^j}{\Delta\tilde{t}}
  $$
- **Dérivée spatiale (ordre 2, centrée) :**
  $$
  \frac{\partial^2 \tilde{H}}{\partial \tilde{z}^2} \approx \frac{\tilde{H}_{k+1} - 2\tilde{H}_k + \tilde{H}_{k-1}}{(\Delta\tilde{z})^2}
  $$

### 1.4. Application du schéma de Crank-Nicolson

Pour assurer la stabilité numérique, nous utilisons un schéma semi-implicite de type Crank-Nicolson. La dérivée spatiale est évaluée comme une moyenne pondérée entre le temps futur ($j+1$) et le temps présent ($j$), contrôlée par le paramètre $\alpha \in [0, 1]$. `pyheatmy` utilise par défaut $\alpha=1$ (schéma totalement implicite).

L'équation discrétisée complète s'écrit :
$$
\gamma_k \frac{\tilde{H}_k^{j+1} - \tilde{H}_k^j}{\Delta\tilde{t}} = \alpha \left( \frac{\tilde{H}_{k+1}^{j+1} - 2\tilde{H}_k^{j+1} + \tilde{H}_{k-1}^{j+1}}{(\Delta\tilde{z})^2} \right) + (1-\alpha) \left( \frac{\tilde{H}_{k+1}^{j} - 2\tilde{H}_k^{j} + \tilde{H}_{k-1}^{j}}{(\Delta\tilde{z})^2} \right) + \zeta_k
$$

### 1.5. Réorganisation en système matriciel : $A \cdot \vec{\tilde{H}}^{j+1} = B \cdot \vec{\tilde{H}}^{j} + \vec{c}$

L'étape cruciale consiste à regrouper tous les termes au temps $j+1$ (les inconnues) à gauche de l'égalité, et tous les termes au temps $j$ (les connus) à droite.

**Termes à gauche (inconnus en $j+1$) :**
$$
-\frac{\alpha}{(\Delta\tilde{z})^2} \tilde{H}_{k-1}^{j+1} + \left(\frac{\gamma_k}{\Delta\tilde{t}} + \frac{2\alpha}{(\Delta\tilde{z})^2}\right) \tilde{H}_{k}^{j+1} -\frac{\alpha}{(\Delta\tilde{z})^2} \tilde{H}_{k+1}^{j+1}
$$

**Termes à droite (connus en $j$) :**
$$
\frac{1-\alpha}{(\Delta\tilde{z})^2} \tilde{H}_{k-1}^{j} + \left(\frac{\gamma_k}{\Delta\tilde{t}} - \frac{2(1-\alpha)}{(\Delta\tilde{z})^2}\right) \tilde{H}_{k}^{j} + \frac{1-\alpha}{(\Delta\tilde{z})^2} \tilde{H}_{k+1}^{j} + \zeta_k
$$

De cette réorganisation, on déduit directement les coefficients des matrices **tridiagonales** $A$ et $B$.

#### Coefficients de la matrice A (implicite, associée à $\vec{\tilde{H}}^{j+1}$)
- **Diagonale principale ($A_{k,k}$):**
  $$
  \text{diag}(A)_k = \frac{\gamma_k}{\Delta\tilde{t}} + \frac{2\alpha}{(\Delta\tilde{z})^2}
  $$
- **Diagonales inférieure et supérieure ($A_{k,k-1}, A_{k,k+1}$):**
  $$
  \text{sub/sup}(A)_k = -\frac{\alpha}{(\Delta\tilde{z})^2}
  $$

#### Coefficients de la matrice B (explicite, associée à $\vec{\tilde{H}}^{j}$)
- **Diagonale principale ($B_{k,k}$):**
  $$
  \text{diag}(B)_k = \frac{\gamma_k}{\Delta\tilde{t}} - \frac{2(1-\alpha)}{(\Delta\tilde{z})^2}
  $$
- **Diagonales inférieure et supérieure ($B_{k,k-1}, B_{k,k+1}$):**
  $$
  \text{sub/sup}(B)_k = \frac{1-\alpha}{(\Delta\tilde{z})^2}
  $$

#### Vecteur C (source et conditions aux limites)
Le vecteur $\vec{c}$ contient le terme de flux latéral $\zeta_k$ ainsi que les termes provenant des conditions aux limites (Neumann), qui modifient les première et dernière lignes des matrices.



## 2. Discrétisation de l'équation de la température ($\tilde{\theta}$)

La méthode de discrétisation est identique à la précédente. La principale différence réside dans la présence d'un terme d'advection qui couple la température au gradient de charge hydraulique.

### 2.1. Rappel de l'équation adimensionnelle

L'équation à résoudre est la suivante :
$$
\beta \cdot \frac{\partial \tilde{\theta}}{\partial \tilde{t}} = \kappa \cdot \frac{\partial \tilde{H}}{\partial \tilde{z}} \cdot \frac{\partial \tilde{\theta}}{\partial \tilde{z}} + \frac{\partial^2 \tilde{\theta}}{\partial \tilde{z}^2}
$$

### 2.2. Approximation des dérivées

Nous utilisons les mêmes approximations par différences finies que précédemment, en ajoutant celle de la dérivée première pour la température :

- **Dérivée temporelle (ordre 1) :**
  $$
  \frac{\partial \tilde{\theta}}{\partial \tilde{t}} \approx \frac{\tilde{\theta}_k^{j+1} - \tilde{\theta}_k^j}{\Delta\tilde{t}}
  $$
- **Dérivée spatiale (ordre 1, centrée) :**
  $$
  \frac{\partial \tilde{\theta}}{\partial \tilde{z}} \approx \frac{\tilde{\theta}_{k+1} - \tilde{\theta}_{k-1}}{2\Delta\tilde{z}}
  $$
- **Dérivée spatiale (ordre 2, centrée) :**
  $$
  \frac{\partial^2 \tilde{\theta}}{\partial \tilde{z}^2} \approx \frac{\tilde{\theta}_{k+1} - 2\tilde{\theta}_k + \tilde{\theta}_{k-1}}{(\Delta\tilde{z})^2}
  $$

Le terme de gradient de charge, $\frac{\partial \tilde{H}}{\partial \tilde{z}}$, est considéré comme **connu**. En effet, dans le schéma de résolution, on calcule d'abord $\vec{\tilde{H}}^{j+1}$ sur tout le domaine, puis on s'en sert pour calculer $\vec{\tilde{\theta}}^{j+1}$. Nous noterons ce gradient connu : $G_k^{j+1} = \left(\frac{\partial \tilde{H}}{\partial \tilde{z}}\right)_k^{j+1}$.

### 2.3. Schéma implicite et réorganisation matricielle

En appliquant un schéma totalement implicite ($\alpha=1$, comme dans `pyheatmy`), tous les termes spatiaux sont évalués au temps futur $j+1$.

L'équation discrétisée complète est :
$$
\beta_k \frac{\tilde{\theta}_k^{j+1} - \tilde{\theta}_k^j}{\Delta\tilde{t}} = \kappa_k \cdot G_k^{j+1} \cdot \left( \frac{\tilde{\theta}_{k+1}^{j+1} - \tilde{\theta}_{k-1}^{j+1}}{2\Delta\tilde{z}} \right) + \left( \frac{\tilde{\theta}_{k+1}^{j+1} - 2\tilde{\theta}_k^{j+1} + \tilde{\theta}_{k-1}^{j+1}}{(\Delta\tilde{z})^2} \right)
$$

Nous regroupons à nouveau les termes inconnus ($\tilde{\theta}^{j+1}$) à gauche et les termes connus ($\tilde{\theta}^{j}$) à droite.

**Termes à gauche (inconnus en $j+1$) :**
$$
\underbrace{\left(-\frac{\kappa_k G_k^{j+1}}{2\Delta\tilde{z}} - \frac{1}{(\Delta\tilde{z})^2}\right)}_{\text{coeff pour } k-1} \tilde{\theta}_{k-1}^{j+1} + \underbrace{\left(\frac{\beta_k}{\Delta\tilde{t}} + \frac{2}{(\Delta\tilde{z})^2}\right)}_{\text{coeff pour } k} \tilde{\theta}_{k}^{j+1} + \underbrace{\left(\frac{\kappa_k G_k^{j+1}}{2\Delta\tilde{z}} - \frac{1}{(\Delta\tilde{z})^2}\right)}_{\text{coeff pour } k+1} \tilde{\theta}_{k+1}^{j+1}
$$

**Termes à droite (connus en $j$) :**
$$
\frac{\beta_k}{\Delta\tilde{t}} \tilde{\theta}_{k}^{j}
$$

On en déduit les coefficients de la matrice $A_{\tilde{\theta}}$. La matrice $B_{\tilde{\theta}}$ est cette fois-ci simplement une matrice diagonale.

#### Coefficients de la matrice A (implicite, associée à $\vec{\tilde{\theta}}^{j+1}$)
- **Diagonale principale ($A_{k,k}$):**
  $$
  \text{diag}(A)_k = \frac{\beta_k}{\Delta\tilde{t}} + \frac{2}{(\Delta\tilde{z})^2}
  $$
- **Diagonale inférieure ($A_{k,k-1}$):**
  $$
  \text{sub}(A)_k = -\frac{\kappa_k G_k^{j+1}}{2\Delta\tilde{z}} - \frac{1}{(\Delta\tilde{z})^2}
  $$
- **Diagonale supérieure ($A_{k,k+1}$):**
  $$
  \text{sup}(A)_k = \frac{\kappa_k G_k^{j+1}}{2\Delta\tilde{z}} - \frac{1}{(\Delta\tilde{z})^2}
  $$

**Note importante :** À cause du terme d'advection, la diagonale inférieure et la diagonale supérieure ne sont pas égales. La matrice $A_{\tilde{\theta}}$ est donc **asymétrique**, contrairement à la matrice $A_{\tilde{H}}$ de la diffusion pure.

#### Coefficients de la matrice B (explicite, associée à $\vec{\tilde{\theta}}^{j}$)
- **Diagonale principale ($B_{k,k}$):**
  $$
  \text{diag}(B)_k = \frac{\beta_k}{\Delta\tilde{t}}
  $$
- **Diagonales inférieure et supérieure :** $0$
